```python
import pandas as pd 
```

```python
import json

with open(r'C:\Users\U1109944\OneDrive - Sanofi\Desktop\TrialGPT\benchmark\data\questions.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

if isinstance(data, list):
    df = pd.json_normalize(data)
elif isinstance(data, dict):
    list_key = next((k for k, v in data.items() if isinstance(v, list)), None)
    df = pd.json_normalize(data[list_key]) if list_key else pd.json_normalize([data])
else:
    raise ValueError("Unsupported JSON structure")

df
```
